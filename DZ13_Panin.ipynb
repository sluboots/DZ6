#### Процесс лемматизации
При помощи библиотеки spacy, можно достаточно легко преобразовать слово в его лемму. Для этого необходимо скачать модель ru_core_news_lg. Есть модели более меньшего размера, но в них намного меньше слов.  
Если не ошибаюсь, модель качается при помощи командной строки в powershell ( у меня windows ). Точно сказать не могу, так как скачивал модель давно, для бакалаврской работы и магистерской диссертации. 

```python
stopwords.words('russian')
nlp = spacy.load('ru_core_news_lg')
nltk.download('stopwords')
```

#### В функции ниже, первым делом слово преобразуется в лемму, затем проходит процесс нормализации, а именно превращая символы в их стандартные формы. Затем удаляются звездочки, лишние новые строчки и лишние пробелы

```python
def clear_text(txt_col):
    rdy_tmp = []
    for txt in txt_col:
        tmp_str = str()
        tmp = []
        doc = nlp(txt)
        for token in doc:
            tmp.append(token.lemma_)
        tmp_str = ' '.join(tmp)
        tmp_str = unicodedata.normalize("NFKD", tmp_str)
        tmp_str = tmp_str.replace('\n', '')
        tmp_str = tmp_str.replace('*', '')
        tmp_str = tmp_str.replace('   ', ' ')
        rdy_tmp.append(tmp_str)
    return rdy_tmp
```

```python
clear_text(norm_text)
```

```python
import pandas as pd
```

```python
counts = Counter(clear_text(norm_text))
top5 = counts.most_common(15)
```

```python
df = pd.DataFrame(top5, columns=['Word', 'Frequency'])
```
