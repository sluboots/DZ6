```python
import pandas as pd
import numpy as np
import psycopg2
```

```python
df = pd.read_csv(r'C:\Users\Maksim\Desktop\datasets\TelecomChurn.csv')
```

```python
group_df = df[['Total day calls', 'Total day minutes']] # Для более удобной группировки, ограничил выборку 2мя столбцами
```

```python
group_df
```

```python
group_df.groupby('Total day calls').agg('min')
```

```python
group_df.groupby('Total day calls').agg('min').apply(np.sum) # Посчитал общую сумму звонков днём. 
```

```python
group_df.transform([np.sqrt, np.exp]) # Интереcных идей не было, по-этому просто извлёк квадратный корень числа и подсчитал экспоненту.
# В результате работы transform, появилась таблица с столбцами, которые я указал в параметрах функции. 
```

```python
group_df['Total day calls'].map(np.sqrt) # В данном случае, функция map, вернула мне Series, где к каждому элементу применили функцию np.sqrt
```
